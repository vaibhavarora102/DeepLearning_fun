```python
import pandas as pd
import os
import shutil
```

```python
# create the data for positive samples

file_path ="covid_chestxray_dataset_master/covid-chestxray-dataset-master/metadata.csv"
images_path="covid_chestxray_dataset_master/covid-chestxray-dataset-master/images"

```

```python
df= pd.read_csv(file_path)
df.head()
```

```python
target_dir="cleaned_dataset_for_covid_project/covid"
if not os.path.exists(target_dir):
    os.mkdir(target_dir)
    print("covid folder created")
else:
    print(" there already exists")
```

```python
cnt = 0

for (i, row) in df.iterrows():
    if row["finding"]=="COVID-19" and row["view"]=="PA":
        filename =row["filename"]
        image_path = os.path.join(images_path, filename)
        image_copy_path = os.path.join(target_dir, filename)
        shutil.copy2(image_path, image_copy_path)
        print("moving image ", cnt)
        cnt+=1
print(cnt)
```
