```python
train_path = "CovidDataset/Train"
val_path = "CovidDataset/Val"
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
# CNN based model
```

```python
model= tf.keras.models.Sequential([
                                   tf.keras.layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(224,224,3)),
                                   tf.keras.layers.Conv2D(64,(3, 3), activation='relu'),#increase non linearity so introduce more parameters
                                   tf.keras.layers.MaxPooling2D(2,2),
                                   tf.keras.layers.Dropout(0.25),# to avoid overfitting

                                   tf.keras.layers.Conv2D(128,(3, 3), activation='relu'),
                                   tf.keras.layers.MaxPooling2D(2,2),
                                   tf.keras.layers.Dropout(0.25),# to avoid overfitting

                                   tf.keras.layers.Conv2D(128,(3, 3), activation='relu'),
                                   tf.keras.layers.MaxPooling2D(2,2),
                                   tf.keras.layers.Dropout(0.25),# to avoid overfitting

                                   tf.keras.layers.Flatten(),
                                   tf.keras.layers.Dense(64,activation='relu'),
  
                                   tf.keras.layers.Dropout(0.25),# to avoid overfitting

                                   tf.keras.layers.Dense(1, activation='sigmoid')

])
```

```python
model.compile(loss=tf.keras.losses.binary_crossentropy, optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
# train from scratch
```

```python
train_datagen= tf.keras.preprocessing.image.ImageDataGenerator(
    rescale = 1./255,
    shear_range= 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
)

test_dataset= tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255)
```

```python
train_generator= train_datagen.flow_from_directory(
    'CovidDataset/Train',
    target_size =(224,224),
    batch_size = 32,
    class_mode ='binary'
)

val_generator= test_dataset.flow_from_directory(
    'CovidDataset/Val',
    target_size=(224,224),
    batch_size = 32,
    class_mode ='binary'
)
```

```python
train_generator.class_indices
```

```python
hist = model.fit_generator(
    train_generator,
    steps_per_epoch=8,
    epochs=10,
    validation_data=val_generator,
    validation_steps = 2
)
```
